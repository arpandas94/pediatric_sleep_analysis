```python
import numpy as np
import pandas as pd

df_sleep = pd.read_csv('Sleep_Data_Final.csv')
df_med = pd.read_csv('med_final.csv')
df_sleep.head(3)
```

```python
df_med.head(3)
```

```python
df_final = pd.merge(df_sleep, df_med, how='inner', left_on = 'Patient_ID', right_on = 'STUDY_PAT_ID')
df_final.head(3)
```

```python
print("Final: ",df_final.shape)
print("Sleep Data:",df_sleep.shape)
print("Med Data:",df_med.shape)
```

```python
df_final = df_final.drop(columns=['STUDY_PAT_ID'])
```

```python
df_final.info()
```

```python
df_final.to_csv("data_sleep_med.csv", index=False)
```
