```python
import os
import pandas as pd
import numpy as np

your_path = 'D:\\St.Clair Data Analytics\\Semester 3\\DAB304-21F-002 Healthcare Analytics 21 Fall - 002\\Final Project\\Data\\Sleep Data\\'
files = os.listdir(your_path)
data = "Patient ID, Eye Movements, Eye Blink, Oral Breathing, Nasal Breathing, Holding Breath, Snoring/Snorting, Body Position: Supine, Body Position: Left, Body Position: Right, Body Position: Prone, Cough, Oxygen Desaturation, Awake, EEG Arousal, Gag, Foot Movement, Movement, Hypopnea, Apnea, Apnea Hypopnea Index, AHI Class\n"
for file in files:
    url_file = "D:\\St.Clair Data Analytics\\Semester 3\\DAB304-21F-002 Healthcare Analytics 21 Fall - 002\\Final Project\\Data\\Sleep Data\\" + file
    f = open(url_file, "r")
    content = f.read()
    file_name = file[:-4]
    eye_mov = content.upper().count('EYE')
    eye_mov = eye_mov + content.upper().count('LOOK')
    eye_blink = content.upper().count('BLINK')
    breath_mouth = content.upper().count('ORAL')
    breath_mouth = breath_mouth + content.upper().count('MOUTH')
    breath_mouth = breath_mouth + content.upper().count('DEEP')
    breath_nose = content.upper().count('NASAL')
    breath_nose = breath_nose + content.upper().count('NOSE')
    breath_hold = content.upper().count('HOLD')
    breath_hold = breath_hold + content.upper().count('OBSTRUCT')
    snoring = content.upper().count('SNOR') #snore/snoring/snorting
    body_supine = content.upper().count('SUPINE')
    body_left = content.upper().count('BODY POSITION: LEFT')
    body_right = content.upper().count('BODY POSITION: RIGHT')
    body_prone = content.upper().count('PRONE')
    cough = content.upper().count('COUGH')
    cough = cough + content.upper().count('CONGEST')
    cough = cough + content.upper().count('NOSE RUN')
    o2_desat = content.upper().count('OXYGEN DESAT')
    awake = content.upper().count('AWAKE')
    eeg_ar = content.upper().count('EEG AROUSAL')
    gag = content.upper().count('GAG')
    foot_mov = content.upper().count('FOOT')
    foot_mov = foot_mov + content.upper().count('LEG')
    mov = content.upper().count('MOVE') #movement/move
    hyp = content.upper().count('HYPOPNEA')
    apn = content.upper().count('APNEA')
    ahi = (hyp+apn)/7 #exact formulae for AHI calculation
    if ahi < 5:
        ahi_class = "NONE"
    elif ahi >= 5 and ahi < 15:
        ahi_class = "NORMAL"
    elif ahi >= 15 and ahi < 30:
        ahi_class = "MILD"
    elif ahi >= 30:
        ahi_class = "SEVERE"
    print(apn,hyp,str(ahi), ahi_class)
    data = data + file_name + "," + str(eye_mov) + "," + str(eye_blink) + "," + str(breath_mouth) + "," + str(breath_nose) + "," + str(breath_hold) + "," + str(snoring) + "," + str(body_supine) + "," + str(body_left) + "," + str(body_right) + "," + str(body_prone) + "," + str(cough) + "," + str(o2_desat) + "," + str(awake) + "," + str(eeg_ar) + "," + str(gag) + "," + str(foot_mov) + "," + str(mov) + "," + str(hyp) + "," + str(apn) + "," + str(ahi) + "," + str(ahi_class) + "\n"

f = open("sleep_data_encoded.csv", "a")
f.write(data)
f.close()
```
