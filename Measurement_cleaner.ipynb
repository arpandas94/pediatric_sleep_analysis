```python
import pandas as pd
import numpy as np

data = pd.read_csv("MEASUREMENT.csv")
data.head(3)
```

```python
data = data.drop(columns=['STUDY_MEAS_ID', 'STUDY_ENC_ID', 'STUDY_PROV_ID', 'MEAS_SOURCE', 'MEAS_RECORDED_DATETIME'])
data.head(3)
```

```python
data
```

```python
data[['BP_Sys','BP_Dia']] = data['MEAS_VALUE_TEXT'].str.split('/',expand=True)
data
```

```python
#data = data.drop(columns=['MEAS_VALUE_TEXT'])
#data['BP_Sys'] = data['BP_Sys'].replace(np.nan, 0)
#data['BP_Dia'] = data['BP_Dia'].replace(np.nan, 0)
data['BP'] = (data['BP_Sys'] + 2*data['BP_Dia'])/3
data = data.drop(columns=['BP_Sys','BP_Dia'])
data.head(10)
```

```python
data.dtypes
```

```python
data['BP_Sys'] = data['BP_Sys'].astype(int)
data['BP_Dia'] = data['BP_Dia'].astype(int)
data.dtypes
```

```python
data.to_csv('measurement.csv')
```

```python
data = pd.read_csv("MEASUREMENT.csv")
data.head(3)
```

```python
data['MEAS_VALUE_NUMBER'] = data['MEAS_VALUE_NUMBER'].replace(np.nan, 0)
data['Meas_val'] = data['MEAS_VALUE_NUMBER'] + data['BP']
data = data.drop(columns=['MEAS_VALUE_NUMBER','BP'])
data
```
